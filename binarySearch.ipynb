```python
def random_array(n=8, low=0, high=100):
    """Return a random list of integers."""
    for i in range(n):
        return random.randint(low,high)


random_array()
```

```python
def binary_search(arr, target):
    """
    Binary search on a sorted list in acsending order.
    """
    low = 0                          # start of the search range
    high = len(arr) - 1              # end of the search range

    while low <= high:
        mid = (low + high) // 2      # middle index
        value = arr[mid]             # value at middle

        if value == target:
            # we found the target
            return mid
        elif value > target:
            # target is smaller, so go left side
            high = mid - 1
        else:
            # value < target → go right side
            low = mid + 1

    # got out of the loop = target not there
    return -1   #Returns -1 if the target is not found
```
