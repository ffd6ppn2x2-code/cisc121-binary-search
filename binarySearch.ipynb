```python
def binary_search(arr, target):
    """
    Basic iterative binary search.
    Returns the index of target if found, otherwise -1.
    Assumes arr is sorted.
    """
    low = 0                      # start of the search range
    high = len(arr) - 1          # end of the search range

    while low <= high:
        mid = (low + high) // 2  # middle index
        value = arr[mid]         # value at the middle

        if value == target:
            # found the target, return its index
            return mid
        elif value > target:
            # target is smaller, so search the left half
            high = mid - 1
        else:
            # target is bigger, so search the right half
            low = mid + 1

    # if we get here, the target is not in the list
    return -1
```

```python
def run_test_case(arr, target, expected_index):
    """Run one test and print result in a simple format."""
    result = binary_search(arr, target)
    passed = (result == expected_index)
    print(f"Array={arr}, target={target}, expected={expected_index}, got={result} -> {'PASS' if passed else 'FAIL'}")


print("=== Binary Search Test Cases ===")

# 1. Target in the middle
run_test_case([1, 3, 5, 7, 9], 7, 3)

# 2. Target at the beginning
run_test_case([2, 4, 6, 8, 10], 2, 0)

# 3. Target not in the list
run_test_case([2, 4, 6, 8, 10], 1, -1)

# 4. Single-element list
run_test_case([5], 5, 0)

# 5. Empty list
run_test_case([], 10, -1)
```
